```python
# World Cup 2026 Office Pool App
#
# This is a first working Streamlit version for an office football pool.
# Participants can enter predictions for every match.
# The admin can enter real results.
# The app calculates points and shows a live leaderboard.
#
# Install:
#   pip install streamlit pandas
#
# Run:
#   streamlit run app.py
#
# Files created automatically:
#   data/matches.csv
#   data/predictions.csv
#   data/results.csv

# Import Path so the app can create and read local data files safely.
from pathlib import Path

# Import pandas for table handling, CSV storage, and leaderboard calculations.
import pandas as pd

# Import Streamlit for the web interface.
import streamlit as st


# Store all app data in a dedicated folder.
DATA_DIR = Path("data")

# Store the match schedule in this file.
MATCHES_FILE = DATA_DIR / "matches.csv"

# Store participant predictions in this file.
PREDICTIONS_FILE = DATA_DIR / "predictions.csv"

# Store real match results in this file.
RESULTS_FILE = DATA_DIR / "results.csv"


# Define the minimum columns needed for the match schedule.
MATCH_COLUMNS = [
    "match_id",
    "stage",
    "group",
    "match_date",
    "home_team",
    "away_team",
]

# Define the columns needed for participant predictions.
PREDICTION_COLUMNS = [
    "participant",
    "match_id",
    "pred_home_goals",
    "pred_away_goals",
]

# Define the columns needed for real match results.
RESULT_COLUMNS = [
    "match_id",
    "real_home_goals",
    "real_away_goals",
]


# Define how many points are awarded for each scoring category.
SCORING_RULES = {
    "correct_outcome": 2,
    "exact_score_bonus": 3,
    "correct_goal_difference_bonus": 1,
    "correct_home_goals_bonus": 1,
    "correct_away_goals_bonus": 1,
}


# Cache data-loading functions so the app remains fast when Streamlit reruns the script.
@st.cache_data
# Load the match schedule from CSV.
def load_matches() -> pd.DataFrame:
    # Create the data directory when it does not exist yet.
    DATA_DIR.mkdir(exist_ok=True)

    # Create a starter match schedule when no schedule exists yet.
    if not MATCHES_FILE.exists():
        # Create a small example schedule that can later be replaced by the official full schedule.
        starter_matches = pd.DataFrame(
            [
                {
                    "match_id": "M001",
                    "stage": "Group stage",
                    "group": "A",
                    "match_date": "2026-06-11",
                    "home_team": "Mexico",
                    "away_team": "South Africa",
                },
                {
                    "match_id": "M002",
                    "stage": "Group stage",
                    "group": "A",
                    "match_date": "2026-06-11",
                    "home_team": "Team A3",
                    "away_team": "Team A4",
                },
            ],
            columns=MATCH_COLUMNS,
        )

        # Save the starter schedule to disk.
        starter_matches.to_csv(MATCHES_FILE, index=False)

    # Load and return the schedule.
    return pd.read_csv(MATCHES_FILE, dtype={"match_id": str})


# Cache the predictions loader so repeated page interactions remain responsive.
@st.cache_data
# Load all saved participant predictions from CSV.
def load_predictions() -> pd.DataFrame:
    # Create the data directory when it does not exist yet.
    DATA_DIR.mkdir(exist_ok=True)

    # Create an empty prediction file when no predictions exist yet.
    if not PREDICTIONS_FILE.exists():
        # Build an empty prediction table with the required columns.
        empty_predictions = pd.DataFrame(columns=PREDICTION_COLUMNS)

        # Save the empty prediction table to disk.
        empty_predictions.to_csv(PREDICTIONS_FILE, index=False)

    # Load and return predictions.
    return pd.read_csv(PREDICTIONS_FILE, dtype={"match_id": str})


# Cache the results loader so admin updates are efficient.
@st.cache_data
# Load all real match results from CSV.
def load_results() -> pd.DataFrame:
    # Create the data directory when it does not exist yet.
    DATA_DIR.mkdir(exist_ok=True)

    # Create an empty results file when no results exist yet.
    if not RESULTS_FILE.exists():
        # Build an empty result table with the required columns.
        empty_results = pd.DataFrame(columns=RESULT_COLUMNS)

        # Save the empty result table to disk.
        empty_results.to_csv(RESULTS_FILE, index=False)

    # Load and return results.
    return pd.read_csv(RESULTS_FILE, dtype={"match_id": str})


# Clear Streamlit caches after writing CSV files.
def refresh_data_cache() -> None:
    # Clear all cached data so the interface shows the latest saved values.
    st.cache_data.clear()


# Save the full predictions table to disk.
def save_predictions(predictions: pd.DataFrame) -> None:
    # Write the predictions table without row numbers.
    predictions.to_csv(PREDICTIONS_FILE, index=False)

    # Clear caches so updated predictions are visible immediately.
    refresh_data_cache()


# Save the full results table to disk.
def save_results(results: pd.DataFrame) -> None:
    # Write the results table without row numbers.
    results.to_csv(RESULTS_FILE, index=False)

    # Clear caches so updated results are visible immediately.
    refresh_data_cache()


# Determine the outcome category for a scoreline.
def get_outcome(home_goals: int, away_goals: int) -> str:
    # Return home win when the home team scored more goals.
    if home_goals > away_goals:
        return "home_win"

    # Return away win when the away team scored more goals.
    if home_goals < away_goals:
        return "away_win"

    # Return draw when both teams scored the same number of goals.
    return "draw"


# Calculate points for one prediction compared with one real result.
def score_prediction(
    pred_home_goals: int,
    pred_away_goals: int,
    real_home_goals: int,
    real_away_goals: int,
) -> int:
    # Start every prediction at zero points.
    points = 0

    # Calculate the predicted match outcome.
    predicted_outcome = get_outcome(pred_home_goals, pred_away_goals)

    # Calculate the real match outcome.
    real_outcome = get_outcome(real_home_goals, real_away_goals)

    # Add points when the participant predicted the correct winner or draw.
    if predicted_outcome == real_outcome:
        points += SCORING_RULES["correct_outcome"]

    # Add exact-score bonus when both predicted scores match the real scores.
    if pred_home_goals == real_home_goals and pred_away_goals == real_away_goals:
        points += SCORING_RULES["exact_score_bonus"]

    # Calculate the predicted goal difference from the home-team perspective.
    predicted_goal_difference = pred_home_goals - pred_away_goals

    # Calculate the real goal difference from the home-team perspective.
    real_goal_difference = real_home_goals - real_away_goals

    # Add a bonus when the goal difference is correct.
    if predicted_goal_difference == real_goal_difference:
        points += SCORING_RULES["correct_goal_difference_bonus"]

    # Add a bonus when the participant predicted the home team's goals correctly.
    if pred_home_goals == real_home_goals:
        points += SCORING_RULES["correct_home_goals_bonus"]

    # Add a bonus when the participant predicted the away team's goals correctly.
    if pred_away_goals == real_away_goals:
        points += SCORING_RULES["correct_away_goals_bonus"]

    # Return the final score for this prediction.
    return points


# Build a detailed scoring table for all participants and completed matches.
def calculate_scores(
    matches: pd.DataFrame,
    predictions: pd.DataFrame,
    results: pd.DataFrame,
) -> pd.DataFrame:
    # Return an empty table when there are no predictions yet.
    if predictions.empty:
        return pd.DataFrame()

    # Return an empty table when there are no real results yet.
    if results.empty:
        return pd.DataFrame()

    # Merge predictions with real results by match ID.
    scored = predictions.merge(results, on="match_id", how="inner")

    # Merge match metadata so the scoring table remains readable.
    scored = scored.merge(matches, on="match_id", how="left")

    # Drop rows where real scores are still missing.
    scored = scored.dropna(subset=["real_home_goals", "real_away_goals"])

    # Convert all goal columns to integers.
    goal_columns = [
        "pred_home_goals",
        "pred_away_goals",
        "real_home_goals",
        "real_away_goals",
    ]

    # Convert each goal column from text/float into integer values.
    for column in goal_columns:
        scored[column] = scored[column].astype(int)

    # Calculate points row by row.
    scored["points"] = scored.apply(
        lambda row: score_prediction(
            row["pred_home_goals"],
            row["pred_away_goals"],
            row["real_home_goals"],
            row["real_away_goals"],
        ),
        axis=1,
    )

    # Return the detailed scoring table.
    return scored


# Build the leaderboard from the detailed scoring table.
def calculate_leaderboard(scored: pd.DataFrame) -> pd.DataFrame:
    # Return an empty table when no scored predictions exist yet.
    if scored.empty:
        return pd.DataFrame(columns=["participant", "points", "matches_scored", "exact_scores"])

    # Identify exact scores for tie-breaking.
    scored["is_exact_score"] = (
        (scored["pred_home_goals"] == scored["real_home_goals"])
        & (scored["pred_away_goals"] == scored["real_away_goals"])
    )

    # Group by participant and calculate total points and tie-break data.
    leaderboard = (
        scored.groupby("participant")
        .agg(
            points=("points", "sum"),
            matches_scored=("match_id", "count"),
            exact_scores=("is_exact_score", "sum"),
        )
        .reset_index()
    )

    # Sort the leaderboard by points first and exact scores second.
    leaderboard = leaderboard.sort_values(
        by=["points", "exact_scores", "matches_scored", "participant"],
        ascending=[False, False, False, True],
    )

    # Add a visible rank number.
    leaderboard.insert(0, "rank", range(1, len(leaderboard) + 1))

    # Return the completed leaderboard.
    return leaderboard


# Render the participant prediction form.
def render_prediction_page(matches: pd.DataFrame, predictions: pd.DataFrame) -> None:
    # Show the page title.
    st.header("Fill in your World Cup predictions")

    # Ask for the participant name.
    participant = st.text_input("Your name")

    # Stop early when the participant has not entered a name.
    if not participant:
        st.info("Enter your name to start filling in predictions.")
        return

    # Select existing predictions for this participant.
    existing = predictions[predictions["participant"].str.lower() == participant.lower()]

    # Create a lookup dictionary for existing predictions.
    existing_lookup = {
        row["match_id"]: row
        for _, row in existing.iterrows()
    }

    # Create a form so all predictions are submitted in one action.
    with st.form("prediction_form"):
        # Store newly entered rows here before saving them.
        new_rows = []

        # Loop over matches in schedule order.
        for _, match in matches.iterrows():
            # Create a readable label for this match.
            label = f'{match["match_id"]}: {match["home_team"]} vs {match["away_team"]} ({match["stage"]})'

            # Show the match label.
            st.markdown(f"**{label}**")

            # Create two columns for home and away score inputs.
            col1, col2 = st.columns(2)

            # Read the participant's existing prediction when available.
            existing_prediction = existing_lookup.get(match["match_id"])

            # Set the default home-goals value.
            default_home = int(existing_prediction["pred_home_goals"]) if existing_prediction is not None else 0

            # Set the default away-goals value.
            default_away = int(existing_prediction["pred_away_goals"]) if existing_prediction is not None else 0

            # Ask for predicted home goals.
            pred_home = col1.number_input(
                f'{match["home_team"]} goals',
                min_value=0,
                max_value=20,
                value=default_home,
                key=f'pred_home_{match["match_id"]}',
            )

            # Ask for predicted away goals.
            pred_away = col2.number_input(
                f'{match["away_team"]} goals',
                min_value=0,
                max_value=20,
                value=default_away,
                key=f'pred_away_{match["match_id"]}',
            )

            # Add this prediction to the rows that will be saved.
            new_rows.append(
                {
                    "participant": participant,
                    "match_id": match["match_id"],
                    "pred_home_goals": pred_home,
                    "pred_away_goals": pred_away,
                }
            )

        # Add a submit button for the full prediction form.
        submitted = st.form_submit_button("Save my predictions")

    # Save predictions after the form is submitted.
    if submitted:
        # Remove this participant's old predictions.
        remaining_predictions = predictions[predictions["participant"].str.lower() != participant.lower()]

        # Convert the newly submitted rows into a dataframe.
        participant_predictions = pd.DataFrame(new_rows, columns=PREDICTION_COLUMNS)

        # Combine old predictions from other people with this participant's new predictions.
        updated_predictions = pd.concat(
            [remaining_predictions, participant_predictions],
            ignore_index=True,
        )

        # Save the updated prediction table.
        save_predictions(updated_predictions)

        # Show a success message.
        st.success("Predictions saved.")


# Render the admin results page.
def render_admin_page(matches: pd.DataFrame, results: pd.DataFrame) -> None:
    # Show the page title.
    st.header("Admin: enter real match results")

    # Create a lookup dictionary for existing results.
    result_lookup = {
        row["match_id"]: row
        for _, row in results.iterrows()
    }

    # Create a form so admin results are saved together.
    with st.form("results_form"):
        # Store entered result rows here before saving them.
        new_results = []

        # Loop over every match.
        for _, match in matches.iterrows():
            # Create a readable label for this match.
            label = f'{match["match_id"]}: {match["home_team"]} vs {match["away_team"]}'

            # Show the match label.
            st.markdown(f"**{label}**")

            # Create columns for score inputs and completion checkbox.
            col1, col2, col3 = st.columns([1, 1, 1])

            # Read the existing result when available.
            existing_result = result_lookup.get(match["match_id"])

            # Decide whether this match already has a saved result.
            has_existing_result = existing_result is not None and pd.notna(existing_result["real_home_goals"])

            # Set default home goals for the admin form.
            default_home = int(existing_result["real_home_goals"]) if has_existing_result else 0

            # Set default away goals for the admin form.
            default_away = int(existing_result["real_away_goals"]) if has_existing_result else 0

            # Ask whether the result should count as completed.
            completed = col3.checkbox(
                "Completed",
                value=has_existing_result,
                key=f'completed_{match["match_id"]}',
            )

            # Ask for real home goals.
            real_home = col1.number_input(
                f'{match["home_team"]} real goals',
                min_value=0,
                max_value=20,
                value=default_home,
                key=f'real_home_{match["match_id"]}',
            )

            # Ask for real away goals.
            real_away = col2.number_input(
                f'{match["away_team"]} real goals',
                min_value=0,
                max_value=20,
                value=default_away,
                key=f'real_away_{match["match_id"]}',
            )

            # Save only completed results.
            if completed:
                new_results.append(
                    {
                        "match_id": match["match_id"],
                        "real_home_goals": real_home,
                        "real_away_goals": real_away,
                    }
                )

        # Add a submit button for the admin form.
        submitted = st.form_submit_button("Save real results")

    # Save results after the admin submits the form.
    if submitted:
        # Convert entered results to a dataframe.
        updated_results = pd.DataFrame(new_results, columns=RESULT_COLUMNS)

        # Save the results table.
        save_results(updated_results)

        # Show a success message.
        st.success("Results saved.")


# Render the leaderboard page.
def render_leaderboard_page(matches: pd.DataFrame, predictions: pd.DataFrame, results: pd.DataFrame) -> None:
    # Show the page title.
    st.header("Live leaderboard")

    # Calculate the detailed scoring table.
    scored = calculate_scores(matches, predictions, results)

    # Calculate the leaderboard.
    leaderboard = calculate_leaderboard(scored)

    # Show a message when no matches have been scored yet.
    if leaderboard.empty:
        st.info("No scored matches yet.")
        return

    # Show the leaderboard.
    st.dataframe(leaderboard, use_container_width=True)

    # Show detailed scoring when the user wants transparency.
    with st.expander("Show detailed scoring"):
        st.dataframe(scored, use_container_width=True)


# Render the rules page.
def render_rules_page() -> None:
    # Show the page title.
    st.header("Scoring rules")

    # Explain the points system.
    st.markdown(
        f"""
        | Category | Points |
        |---|---:|
        | Correct outcome, win/draw/loss | {SCORING_RULES['correct_outcome']} |
        | Exact score bonus | {SCORING_RULES['exact_score_bonus']} |
        | Correct goal difference bonus | {SCORING_RULES['correct_goal_difference_bonus']} |
        | Correct home-team goals bonus | {SCORING_RULES['correct_home_goals_bonus']} |
        | Correct away-team goals bonus | {SCORING_RULES['correct_away_goals_bonus']} |

        Example: if the real result is 2–1 and someone predicted 3–2, they get the correct outcome and correct goal difference.
        """
    )


# Define the main app entry point.
def main() -> None:
    # Configure the Streamlit browser tab and layout.
    st.set_page_config(page_title="World Cup 2026 Office Pool", layout="wide")

    # Show the app title.
    st.title("World Cup 2026 Office Pool")

    # Load the match schedule.
    matches = load_matches()

    # Load participant predictions.
    predictions = load_predictions()

    # Load real results.
    results = load_results()

    # Create sidebar navigation.
    page = st.sidebar.radio(
        "Choose page",
        ["Predictions", "Leaderboard", "Admin results", "Rules"],
    )

    # Show the prediction page.
    if page == "Predictions":
        render_prediction_page(matches, predictions)

    # Show the leaderboard page.
    elif page == "Leaderboard":
        render_leaderboard_page(matches, predictions, results)

    # Show the admin page.
    elif page == "Admin results":
        render_admin_page(matches, results)

    # Show the rules page.
    elif page == "Rules":
        render_rules_page()


# Run the app only when this file is executed directly.
if __name__ == "__main__":
    # Start the Streamlit app.
    main()
```
